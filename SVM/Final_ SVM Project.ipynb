# One Hot Encoding 

```python
#converting type of columns to category 
df['Race'] = df['Race'].astype('category') 
df['Marital Status'] = df['Marital Status'].astype('category')
df['T Stage '] = df['T Stage '].astype('category') 
df['N Stage'] = df['N Stage'].astype('category') 
df['6th Stage'] = df['6th Stage'].astype('category') 
df['differentiate'] = df['differentiate'].astype('category') 
df['A Stage'] = df['A Stage'].astype('category') 
df['Estrogen Status'] = df['Estrogen Status'].astype('category') 
df['Progesterone Status'] = df['Progesterone Status'].astype('category') 
```

```python
#assigning numerical values and storing it in another column
df['Race_new'] = df['Race'].cat.codes 
df['Marital Status_new'] = df['Marital Status'].cat.codes 
df['T Stage new'] = df['T Stage '].cat.codes 
df['N Stage_new'] = df['N Stage'].cat.codes 
df['6th Stage_new'] = df['6th Stage'].cat.codes 
df['differentiate_new'] = df['differentiate'].cat.codes 
df['A Stage_new'] = df['A Stage'].cat.codes 
df['Estrogen Status_new'] = df['Estrogen Status'].cat.codes 
df['Progesterone Status_new'] = df['Progesterone Status'].cat.codes 
```

```python
#one hot enocoder function definition
enc = OneHotEncoder(drop='first')
```

```python
#pass encoded columns
enc_data = pd.DataFrame(enc.fit_transform(df[['Race_new', 'Marital Status_new', 'T Stage new', 'N Stage_new', '6th Stage_new', 'differentiate_new', 'A Stage_new', 'Estrogen Status_new', 'Progesterone Status_new']]).toarray())
```

```python
new_df = df.join(enc_data)
new_df
```

Feature Selection

```python
#select inputs and output
x = df[['Race_new','Marital Status_new', 'T Stage new', 'N Stage_new', '6th Stage_new', 'differentiate_new', 'A Stage_new', 'Estrogen Status_new', 'Progesterone Status_new']]
y = df['Status']
```

```python
#split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.2,random_state=42)
```

```python
#declare SVM model 
model=svm.SVC()
```

```python
#fit model 
model.fit(X_train, y_train)
```

```python
#predict 
predictions = model.predict(X_test)
```
