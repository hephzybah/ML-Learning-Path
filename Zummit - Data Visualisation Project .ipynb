```python
#import libraries 
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("Phone_Classification.csv")
df.head(20)
```

```python
df.columns
```

# Classification of Cores - Pie Chart

```python
#extract column 'cores'
core = df['n_cores']
core
```

```python
#count the occurence of cores per core 
n_core_count = core.value_counts()
n_core_count
```

```python
#plot pie chart 
plt.pie(n_core_count, autopct='%1.1f%%', startangle=140)
plt.title('Percentage of Cores Avaiable in the Phone Classification Dataset')
plt.axis('equal')
plt.show()
```

# Line Chart - Clockspeed 

```python
#take out the clock speed parameter 
clock_speed = df['clock_speed']
clock_speed
```

```python
#turn it into a dataframe and rename the quantity column
df_clock = pd.DataFrame(clock_speed)
df_clock.columns = ['Quantity']

#turn the index into a data column 
df_reset = df_clock.reset_index()
df_reset
```

```python
#check the length of the new index name and modify it accordingly 
'''if len(df_reset.columns) == 2:
    df_reset.columns = ['Clock Speed', 'Quantity']
df_reset'''
```

```python
#create scatter plot 
x = df_reset['index']
y = df_reset['Quantity']

plt.scatter(x, y, color='green', marker = 'x', label='Scatter Plot')

plt.xlabel('Clock Speed')
plt.ylabel('Quantity')
plt.title('Scatter Plot Showing ClockSpeeds of Different Phones')

plt.legend()

plt.show()
```

# Bar Chart - Price Range

```python
wifi = df['wifi'].value_counts()
wifi
```

```python
df_wifi = pd.DataFrame(wifi)
df_wifi
```

```python
df_wifi.columns = ['Wifi']

#turn the index into a data column 
df_reset_wifi = df_wifi.reset_index()
df_reset_wifi
```

```python
xx = df_reset_wifi['index']
yy = df_reset_wifi['Wifi']

plt.bar(x, y, color='blue', alpha=0.6)

plt.xlabel('Wifi')
plt.ylabel('Quantity')
plt.title('Bar Plot of Phones with Wifi')


plt.show()
```
